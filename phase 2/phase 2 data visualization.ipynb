# cleaning the data

```python
import pandas as pd
```

```python
df = pd.read_csv('iati-activities-with-location.csv')
```

```python
df.head()
```

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
df['description'].value_counts().idxmax()
```

```python
#replace the NAN by the most frequent 
df["description"].replace(np.nan, 'UNICEF Umbrella Grant to implement activities designed to improve MCH through a wide range of interventions that may include Integrated Management of Childhood Illness (IMCI); micronutrient fortification and supplementation programs (specifically iodine and vitamin A).', inplace = True)
```

```python
#checking null values
df.isnull().sum()
```

```python
df['location_name'].value_counts().idxmax()
```

```python
#replace the NAN by the most frequent 
df["location_name"].replace(np.nan,'Pakistan', inplace = True)
```

```python
df.isnull().sum()
```

```python
df['sector_code'].value_counts().idxmax()
```

```python
#replace the NAN by the most frequent 
df["sector_code"].replace(np.nan,'Democratic participation and civil society', inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(index=df.index[0], 
        axis=0, 
        inplace=True)
```

```python
df.head()
```

```python
df.to_csv('cleaned_iati.csv')
```

```python
#Reading the dataset
test = pd.Series(-1 + 2 * np.random.rand(10), index=list('abcdefghij'))
#Function for makig a waterfall chart
def waterfall(series):
    df = pd.DataFrame({'pos':np.maximum(series,0),'neg':np.minimum(series,0)})
    blank = series.cumsum().shift(1).fillna(0)
    df.plot(kind='bar', stacked=True, bottom=blank, color=['r','b'], alpha=0.5)
    step = blank.reset_index(drop=True).repeat(3).shift(-1)
    step[1::3] = np.nan
    plt.plot(step.index, step.values,'k')
#Creating the waterfall chart
waterfall(test)
#Adding the aesthetics
plt.title('aid project')
plt.xlabel('latitude')
plt.ylabel('longitude')
#Show the plot
plt.show()
```
