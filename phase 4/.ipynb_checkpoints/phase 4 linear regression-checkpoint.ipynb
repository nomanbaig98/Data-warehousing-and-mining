# cleaning the data

```python
import pandas as pd
```

```python
df = pd.read_csv('iati-activities-with-location.csv')
```

```python
df.head()
```

```python
import numpy as np

# replace "?" to NaN
df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("") 
```

```python
df['description'].value_counts().idxmax()
```

```python
#replace the NAN by the most frequent 
df["description"].replace(np.nan, 'UNICEF Umbrella Grant to implement activities designed to improve MCH through a wide range of interventions that may include Integrated Management of Childhood Illness (IMCI); micronutrient fortification and supplementation programs (specifically iodine and vitamin A).', inplace = True)
```

```python
#checking null values
df.isnull().sum()
```

```python
df['location_name'].value_counts().idxmax()
```

```python
#replace the NAN by the most frequent 
df["location_name"].replace(np.nan,'Pakistan', inplace = True)
```

```python
df.isnull().sum()
```

```python
df['sector_code'].value_counts().idxmax()
```

```python
#replace the NAN by the most frequent 
df["sector_code"].replace(np.nan,'Democratic participation and civil society', inplace = True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(index=df.index[0], 
        axis=0, 
        inplace=True)
```

```python
df.head()
```

```python
df.to_csv('cleaned_iati.csv')
```

# implementing linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
import pandas as pd
df1 = pd.read_csv('cleaned_iati.csv')
df1.shape
df1.describe
```

```python
df1.plot(x='location_latitude', y='location_longitude', style='o')
plt.title('latitude vs longitude')
plt.xlabel('Min length')
plt.ylabel('Max length')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.tight_layout()
seabornInstance.distplot(df1['location_latitude'])
```

```python
X = df1['location_latitude'].values.reshape(-1,1)
y = df1['location_longitude'].values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=0)

regressor = LinearRegression()

regressor.fit(X_train, y_train) #training the algorithm

#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)

y_pred = regressor.predict(X_test)

df = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted':

y_pred.flatten()})

df1
```

```python
plt.scatter(X_test, y_test,
color='gray')
plt.plot(X_test, y_pred,
color='red', linewidth=2)
plt.show()
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test,
y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test,
y_pred))
print('Root Mean Squared Error:',
np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
